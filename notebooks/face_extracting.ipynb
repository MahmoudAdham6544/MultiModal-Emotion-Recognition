```python
import cv2
import os
output_parent_dir = r'../data/Processed/Video_Actor_Faces'
os.makedirs(output_parent_dir, exist_ok=True)
# Load the Haar cascade file for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
actors_path = r'../data/Unprocessed/Video_Actors'
for actor in os.listdir(actors_path):
    os.makedirs(output_parent_dir +'/'+ actor, exist_ok=True) #Create an output directory for each actor
    for actor_sub in os.listdir(actors_path +'/'+ actor):
        for video in os.listdir(actors_path +'/'+ actor +'/'+ actor_sub):
            output_dir = output_parent_dir + '/' + actor + '/' + video[:-4] #Remove the .mp4 extension
            os.makedirs(output_dir, exist_ok=True) #Create an output directory for each video
            video_path = actors_path +'/'+ actor +'/'+ actor_sub +'/'+ video
            cap = cv2.VideoCapture(video_path)

            frame_count = 0
            frame_skip = 2 # Number of frames to skip between face extractions

            while cap.isOpened():
                ret, frame = cap.read()
                if not ret:
                    break

                frame_count += 1
                if frame_count % frame_skip != 0:
                    continue
                gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

                # Detect faces
                faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

                for (x, y, w, h) in faces:
                    face = frame[y:y+h, x:x+w]
                    face = cv2.resize(face, (224, 224), interpolation=cv2.INTER_AREA) # Resize the face to 224x224 pixels
                    face_filename = os.path.join(output_dir, f"face_{frame_count}.jpg")
                    cv2.imwrite(face_filename, face)

            cap.release()
            print("Done extracting faces.")
```
